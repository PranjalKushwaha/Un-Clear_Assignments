Importing required libraries.

```python
import tensorflow as tf
import keras
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
print(tf.__version__)
```

Reading training data from disk as a pandas dataframe.

```python
#train_path = keras.utils.get_file("digit_train","https://drive.google.com/file/d/1xRYeULNBxplVhOEP53-kLg7xWfH7CA4x/view?usp=sha0ring",)
train_path = "/home/pranjal/Desktop/train.csv"
train = pd.read_csv(train_path)
```

Separating data into train and test batches and removing labels.

```python
tra = train.sample(38000)
tl = tra.pop('label')
tra = (tra/128)-1
test = train.drop(tra.index)
tel = test.pop('label')
test = (test/128)-1
```

Creating the model.

```python
model = keras.Sequential(
    [
        keras.layers.Dense(512,activation= tf.nn.relu,input_shape = (1,784)),
        keras.layers.Dense(256,activation = tf.nn.relu),
        keras.layers.Dense(128,activation = tf.nn.relu),
        keras.layers.Dense(50, activation = tf.nn.relu),
        keras.layers.Dropout(rate = 0.3),
        keras.layers.Dense(units = 10, activation = tf.nn.softmax)
    ])
model.compile(optimizer=tf.optimizers.Adam(lr = 0.002),
             metrics = ['accuracy'],
             loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True))
```

```python
model.summary()
```

Training the model for 300 epochs.

```python
history = model.fit(x = tra,
                    y = tl,
                    epochs=300,
                    shuffle=True,
                    batch_size = 3800,
                    validation_split = 0.2)
```

Saving the trained model to the disk.

```python
model.save('home/Desktop/Model')
```

Evaluating the model accuracy on the test dataset.

```python
model.evaluate(x=test,y= tel)
```

Plotting the accuracy vs epoch graph.

```python
hist = pd.DataFrame(history.history)
epochs = history.epoch
def plot_curve(epochs, hist, list_of_metrics):
  plt.figure()
  plt.xlabel("Epoch")
  plt.ylabel("Value")

  for m in list_of_metrics:
    x = hist[m]
    plt.plot(epochs[1:], x[1:], label=m)

  plt.legend()
list = ['accuracy'] 
plot_curve(epochs = epochs, hist = hist, list_of_metrics = list)
```

Predicting the results on the unlabelled dataset.

```python
t = pd.read_csv('/home/pranjal/Desktop/test.csv')
t = (t/128)-1
```

```python
#model = tf.keras.models.load_model('home/Desktop/Model')
x = model.predict(t)
```

Writing the predicted values into a csv file.

```python
a = open("submission.csv", "w")

for i in range(len(x)):
    max = 0;
    for j in range(10):
        if x[i][j] > x[i][max]:
            max = j
    print('%d,%d'%(i+1,max),file =a )


a. close()
```
